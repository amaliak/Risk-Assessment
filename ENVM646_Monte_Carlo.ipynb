```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Q1 =np.random.normal(15, 2, 365)
# plotting the points  
plt.plot(Q1)

# naming the x axis 
plt.xlabel('Day') 
# naming the y axis 
plt.ylabel('Flow $(m^3/day)$') 
  
# giving a title to my graph 
plt.title('Flow realizations in modeled creek measured once a day') 
  
# function to show the plot 
plt.show() 
```

```python
C1 =np.random.normal(0.5, 0.02, 365)
# plotting the points  
plt.plot(C1)

# naming the x axis 
plt.xlabel('Day') 
# naming the y axis 
plt.ylabel('Concentration $(mg/L)$') 
  
# giving a title to my graph 
plt.title('Concentrations in modeled creek measured once a day') 
  
# function to show the plot 
plt.show() 
```

```python
plt.hist(C1)
plt.title("Concentration Histogram")
plt.xlabel("Concentration(mg/L)")
plt.ylabel("Frequency")
```

```python
plt.hist(Q1)
plt.title("Q1 Histogram")
plt.xlabel("Concentration(mg/L)")
plt.ylabel("Frequency")
```

```python
# Calculate output flow
Q2 = 2  # m3/sec
C2 = 8 # mg/L

Qout = Q1 + Q2; 

plt.hist(Qout)
plt.title("Qout Histogram")
plt.xlabel("Qout$(m^3/day)$")
plt.ylabel("Frequency")
```

```python
import pandas as pd
# Calculate output concentration
Cout = (Q1*C1 + Q2*C2)/Qout
df = pd.DataFrame({'Q1':Q1, 'Qout': Qout, 'C1':C1, 'Cout':Cout})
```
